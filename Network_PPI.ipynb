```python
import networkx as nx
import matplotlib.pyplot as plt
from pyvis.network import Network
from IPython.core.display import display, HTML
```

```python
with open('/home/jairlinux/Documents/UPY/8th/SNA/practice-sna/networks/protein.edgelist.txt', 'r') as file:
    edges = [line.strip().split() for line in file.readlines()]
print(edges)
```

```python
G = nx.Graph()

# Add edges to the graph
G.add_edges_from(edges)

print(G.edges())
```

```python
# Draw the graph with matplotlib
plt.figure(figsize=(40, 40))
pos = nx.spring_layout(G)  # You can experiment with different layouts
nx.draw(G, pos, with_labels=True, node_size=60, font_size=5, edge_color='gray', node_color='skyblue', font_color='black', width=0.5)

# Show the plot
plt.show()
```

```python
degree_centrality = nx.degree_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)
katz_centrality = nx.katz_centrality_numpy(G)
pagerank = nx.pagerank(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
```

```python
# Print top 5 nodes for each centrality measure
print("Top 5 nodes by Degree Centrality:", sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5])
print("Top 5 nodes by Eigenvector Centrality:", sorted(eigenvector_centrality.items(), key=lambda x: x[1], reverse=True)[:5])
print("Top 5 nodes by Katz Centrality:", sorted(katz_centrality.items(), key=lambda x: x[1], reverse=True)[:5])
print("Top 5 nodes by PageRank:", sorted(pagerank.items(), key=lambda x: x[1], reverse=True)[:5])
print("Top 5 nodes by Betweenness Centrality:", sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:5])
print("Top 5 nodes by Closeness Centrality:", sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:5])
```

```python
# Visualization (e.g., for Degree Centrality)
plt.figure(figsize=(20, 20))
pos = nx.spring_layout(G)
node_size = [v * 10000 for v in degree_centrality.values()]
nx.draw_networkx(G, pos, node_size=node_size, with_labels=False, node_color=list(degree_centrality.values()), cmap=plt.cm.viridis)
plt.title("Network Visualization with Degree Centrality")
plt.show()
```

```python
# Get degree values
degree_values = list(degree_centrality.values())

# Plot degree distribution
plt.figure(figsize=(10, 6))
plt.hist(degree_values, bins=30, edgecolor='black')
plt.title("Degree Distribution of the PPI Network")
plt.xlabel("Degree Centrality")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Identify the largest connected component
largest_cc = max(nx.connected_components(G), key=len)
G_lcc = G.subgraph(largest_cc).copy()

# Measure network characteristics for the largest connected component
size_of_network = G_lcc.number_of_nodes()
number_of_links = G_lcc.number_of_edges()
average_path_length = nx.average_shortest_path_length(G_lcc)
clustering_coefficient = nx.average_clustering(G_lcc)

# Calculate distance metrics for the largest connected component
average_distance = nx.average_shortest_path_length(G_lcc)
diameter = nx.diameter(G_lcc)
eccentricity = nx.eccentricity(G_lcc)
radius = nx.radius(G_lcc)
periphery = list(nx.periphery(G_lcc))
center = list(nx.center(G_lcc))

# Print the results
print("Size of the Network:", size_of_network)
print("Number of Links:", number_of_links)
print("Average Path Length:", average_path_length)
print("Clustering Coefficient:", clustering_coefficient)
print("Average Distance:", average_distance)
print("Diameter:", diameter)
print("Radius:", radius)
print("Periphery:", periphery)
print("Center:", center)
```
